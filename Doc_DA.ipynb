```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("processed_data.csv")
```

```python
print("\n UNIQUE VALUES ")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
print(df.columns.tolist())
```

```python
print("\n TOP 5 DOCTORS BY FEEDBACK COUNT ")
print(df.sort_values("feedback_count", ascending=False).head())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["avg_feedback_score"], kde=True, bins=15)
plt.title("Distribution of Average Feedback Scores")
plt.xlabel("Average Feedback Score")
plt.ylabel("Doctor Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["feedback_count"], kde=False, bins=15)
plt.title("Distribution of Feedback Counts")
plt.xlabel("Number of Feedbacks")
plt.ylabel("Doctor Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_doctors = df.sort_values("avg_feedback_score", ascending=False).head(10)
sns.barplot(x="avg_feedback_score", y="doctor_name", data=top_doctors)
plt.title("Top 10 Doctors by Average Feedback Score")
plt.xlabel("Average Feedback Score")
plt.ylabel("Doctor Name")
plt.show()
```
